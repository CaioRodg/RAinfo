Estarei tratando aqui a relação entre **sale_amount** X **comissões**.

Logo abaixo realizei a criação de um novo "DataFrame" com base no primeiro, porém apenas com as informações do nome; Vendas Líquidas; Comissões Líquidas

```python
relat = dfra[['site_name', 'amount_sold', 'sale_amount','comissão']]
relat = pd.DataFrame(data=relat)
relat = relat.set_index('site_name')
relat
```

Devido ao alto número de duplicidades, realizei união através do groupby e uma somatória para que possamos visualizar melhor.

```python
values = relat.groupby('site_name').sum()
values
```

```python
values.sort_values(by=['comissão'])
```

```python
values['Resultado %'] = values.comissão.div(values.sale_amount)*100
values.sort_values(by=['Resultado %']).round(2)
```

Levando em consideração que a *comissão é o valor pago ao vendedor* que efetuou a venda, analisei que as redes que aumentariam o faturamento são as que pagam menos comissão. Fiz um cálculo do % de comissão por rede e listei as que tinham o menor percentual. Entendo que elas conseguiriam aumentar o valor total de faturamento mais rápido.

São elas:

1.   **C&C BR**
2.   **SorteOnline BR**
